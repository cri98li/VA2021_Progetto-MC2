```python
date = []
time = []

for i in cc_data.timestamp:
    splitted = i.split(" ")
    date.append(splitted[0])
    time.append(splitted[1])
    
cc_data.timestamp = date
cc_data['time'] = time

cc_data
```

```python
cc_data['credit_card'] = True
loyalty_data['loyalty_card'] = True

cards_data = pd.merge(cc_data,
                  loyalty_data,
                  on=['price','timestamp', 'location'],
                  how='outer',
                  indicator=True)

cards_data['credit_card'].fillna(False, inplace=True)
cards_data['loyalty_card'].fillna(False, inplace=True)
```

# Cosa ho scoperto

Ci sono 3 insiemi di dati che contengono FirstName e LastName. cc_data e loyalty_data sono congiunti ma non uguali, car è contenuto nell'intersezione dei due.
- Tutti gli utenti che hanno noleggiato un'auto hanno sia la carta di credito che la carta fedeltà
- Non tutti gli utenti hanno noleggiato un'auto
- Non tutti gli utenti che pagano con la carta di credito hanno la carta fedeltà
- Non tutti gli utenti che hanno la carta fedeltà pagano con la carta di credito



Si nota inoltre che car non contiene gli id di tutte le auto tracciate in GPS.

## Cosa farò
1. Creo un nuovo csv utenti con la lista di tutti gli utenti e un relativo id. Questa lista conterrà inoltre un utente finto per ogni macchina tracciata che non è presente in car assignment 
2. Sostituisco nei tre insiemi il nome e il cognome con l'id (occupo meno spazio)
3. Creo un df con le transazioni unificate ed aggiungo le colonne: "credit_card" e "loyalty_card"

```python
nomi = car[['FirstName', 'LastName', 'CurrentEmploymentType', 'CurrentEmploymentTitle', 'CarID']].drop_duplicates().reset_index(drop=True)

nomi = pd.merge(nomi, gps["CarID"].drop_duplicates(), on="CarID", how='outer')

nomi['id'] = nomi.index

macchine = nomi.fillna(0) #NOTA: macchina con id = 0 --> non esiste

macchine = macchine[["id", "CarID"]].astype(int)

macchine.to_csv("macchine.csv", index=False)



nomi.drop(columns="CarID", inplace=True)

nomi.to_csv("nomi.csv", index=False)

nomi
```
